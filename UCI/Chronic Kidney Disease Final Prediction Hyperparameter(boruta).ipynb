# SMOTE Technique

```python
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import SVMSMOTE 
```

```python
smote=SMOTE()
sm = SVMSMOTE()
```

```python
x_train_resample,y_train_resample=sm.fit_resample(u,v) 
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
k=scaler.fit_transform(x_train_resample)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(k,y_train_resample,test_size=0.2, random_state=0)
```

```python
from sklearn.model_selection import ShuffleSplit
fold=ShuffleSplit(n_splits=5, test_size=0.1, random_state=0)
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RandomizedSearchCV
```

# Support Vector Machine

```python
from sklearn.svm import SVC
svm=SVC()
param_grid={'C':[0.1,1,10,100],'gamma':['scale'],'kernel':['rbf']}
svm_randomgrid=RandomizedSearchCV(estimator=svm,param_distributions=param_grid,verbose=1,n_jobs=-1,cv=fold)
svm_randomgrid.fit(x_train,y_train)
svm_randomgrid.best_estimator_, svm_randomgrid.best_score_
```

# K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
param_grid={'n_neighbors':range(1, 21, 2),
            'weights':['uniform', 'distance'],
            'metric':['euclidean', 'manhattan', 'minkowski']}
knn_randomgrid=RandomizedSearchCV(estimator=knn,param_distributions=param_grid,verbose=4,n_jobs=6,cv=fold)
knn_randomgrid.fit(x_train,y_train)
knn_randomgrid.best_params_, knn_randomgrid.best_score_
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gdb=GradientBoostingClassifier()
param_grid={'learning_rate':[0.01,0.05,0.1],
            'n_estimators':[100,200,300,400,500],
            'max_depth':[2,3,4,5,6,7,8],
            'min_samples_leaf':[4,8,10]}
gdb_randomgrid=RandomizedSearchCV(estimator=gdb,param_distributions=param_grid,verbose=1,n_jobs=-1,cv=fold)
gdb_randomgrid.fit(x_train,y_train)
gdb_randomgrid.best_params_, gdb_randomgrid.best_score_
```

# LightGBM

```python
from lightgbm import LGBMClassifier
lgb=LGBMClassifier()
param_grid={'learning_rate':[0.05,0.1],
            'n_estimators':[100,200,300,400,500],
            'max_depth':[2,3,4,5,6,7,8,9,10],
            'min_child_weight':[1,3,5,7,9]}
lgb_randomgrid=RandomizedSearchCV(estimator=lgb,param_distributions=param_grid,verbose=True,n_jobs=4,cv=fold)
lgb_randomgrid.fit(x_train,y_train)
lgb_randomgrid.best_params_, lgb_randomgrid.best_score_
```

# XgBoost

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
param_grid={'learning_rate':[0.05,0.1,0.15,0.2,0.25],
            'n_estimators':[100,200,300,400,500,600,700],
            'max_depth':[2,3,4,5,6,7,8,9,10],
            'min_child_weight':[1,2,3,4,5,6,7,8,9],
            'eval_metric':['error']}
xgb_randomgrid=RandomizedSearchCV(estimator=xgb,param_distributions=param_grid,verbose=True,n_jobs=4,cv=fold)
xgb_randomgrid.fit(x_train,y_train)
xgb_randomgrid.best_params_, xgb_randomgrid.best_score_
```
