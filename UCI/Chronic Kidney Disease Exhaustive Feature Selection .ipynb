# Exhaustive Feature Selection

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector
from sklearn.linear_model import LogisticRegression as LGR
```

```python
feature_names=tuple(u.columns)
feature_names
```

```python
efs=ExhaustiveFeatureSelector(LGR(max_iter=20),min_features=1,max_features=14,scoring='accuracy',n_jobs=8,cv=3)
```

```python
efs.fit(k,v,custom_feature_names=feature_names)
```

```python
best_feature=efs.best_feature_names_
```

```python
best_feature
```

```python
df=pd.DataFrame.from_dict(efs.get_metric_dict()).T
df=df[['feature_idx','avg_score']]
```
