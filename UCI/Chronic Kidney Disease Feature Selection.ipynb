# Feature Selection-Boruta

```python
from boruta import BorutaPy
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
feat_selector=BorutaPy(rf,n_estimators=200,verbose=2, random_state=None)
```

```python
u=a.iloc[:,:-1]
v=a['Label']
```

```python
u_shadow=u.apply(np.random.permutation)
u_shadow.columns=['shadow_'+feat for feat in u.columns]
u_boruta=pd.concat([u,u_shadow],axis=1)
```

```python
u_boruta
```

```python
#from sklearn.preprocessing import StandardScaler
#scaler=StandardScaler()
#k=scaler.fit_transform(u_boruta)
```

```python
rf.fit(u_boruta,v)
```

```python
feature_imp_u=rf.feature_importances_[:len(u.columns)]
feature_imp_shadow=rf.feature_importances_[len(u.columns):]
hits=feature_imp_u>feature_imp_shadow.max()
```

```python
feature_imp_u
```

```python
feature_imp_shadow.max()
```

```python
hits
```

```python
feature_names = np.array(u.columns)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
k=scaler.fit_transform(u)
```

```python
feat_selector.fit(k,v)
```

```python
feature_ranks = list(zip(feature_names, 
                         feat_selector.ranking_, 
                         feat_selector.support_))
```

```python
for feat in feature_ranks:
    print('Feature: {:<30} Rank: {},  Keep: {}'.format(feat[0], feat[1], feat[2]))
```

```python
feat_selector.n_features_
```

```python
#Select best feature automatically
```

```python
best_feature=feat_selector.transform(k)
best_feature
```

# Recursive Feature Elimination Method

```python
from sklearn.feature_selection import RFE
sel=RFE(RandomForestClassifier())
```

```python
sel.fit(k,v)
```

```python
sel.get_support()
```

```python
features=u.columns[sel.get_support()]
features
```

```python
for i in range(u.shape[1]):
	print('Column: %d, Selected %s, Rank: %.3f' % (i, sel.support_[i], sel.ranking_[i]))
```
